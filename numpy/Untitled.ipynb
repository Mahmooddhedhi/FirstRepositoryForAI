# easy to use (simplicity)

```python
l = [[1,2,3,5],
    [4,5,6,5],
    [7,8,9,5]]

l
```

```python
for rid,row in enumerate(l):
    print(rid,row)
    for cid,_ in enumerate(row):
        print(cid, end="***")
    print()
```

```python
for rid,row in enumerate(l):
    for cid,_ in enumerate(row):
        l[rid][cid] += 5
l
```

# now apply same above operation with numpy

```python
l1 = np.array([[1,2,3,5],
    [4,5,6,5],
    [7,8,9,5]])
l1
```

```python
l1 + 5
```
